```python
# check if all nrrds available
list_past = []
list_error = []
for subdir in os.listdir(dir):
    if ".zip" not in subdir:
        for manifest_dir in os.listdir(os.path.join(dir, subdir)):
            if "Manifest" in manifest_dir:
                for work_space_name in os.listdir(os.path.join(dir, subdir,manifest_dir)):
                    dirs = os.listdir(os.path.join(dir, subdir,manifest_dir, work_space_name))
                    dirs.sort()
                    for scan_sessions in dirs: 
                        if ".json" not in scan_sessions:
                            print("Now Processing: " + str(scan_sessions))
                            ct_nrrd = False
                            pt_nrrd = False                            
                            for nrrd_files in os.listdir(os.path.join(dir, subdir,manifest_dir, work_space_name, scan_sessions)): 
                                for file in os.listdir(os.path.join(dir, subdir,manifest_dir, work_space_name, scan_sessions, nrrd_files)): 
                                    if ".json" not in file and "ct" in file:
                                        ct_nrrd = True
                                    if ".json" not in file and "pt" in file:
                                        pt_nrrd = True
                                if ct_nrrd and pt_nrrd:
                                    print("Past")
                                    list_past.append(scan_sessions)
                                else:
                                    print(str(scan_sessions) + " missing nrrd, please check")
                                    list_error.append(scan_sessions)
```
